```python
def song_ids(user, playlist_id):
    ids = []
    playlist = sp.user_playlist(user, playlist_id)
    for item in playlist['tracks']['items']:
        track = item['track']
        ids.append(track['id'])
    return ids
```

```python
def song_features(id):
    meta = sp.track(id)
    features = sp.audio_features(id)

    # meta
    name = meta['name']
    artist = meta['album']['artists'][0]['name']

    # features
    acousticness = features[0]['acousticness']
    danceability = features[0]['danceability']
    energy = features[0]['energy']
    instrumentalness = features[0]['instrumentalness']
    liveness = features[0]['liveness']
    loudness = features[0]['loudness']
    speechiness = features[0]['speechiness']
    valence = features[0]['valence']

    track = [name, artist,acousticness, danceability, energy, instrumentalness, liveness, loudness, speechiness, valence]
    return track
```

```python
def recommend_songs(track_id, df, n_recs = 5):
    
    ref_df = df.copy()
    
    # get danceability and energy of given track from spotify api
    track_features = song_features(track_id)
    track_songvec = np.array([track_features[3], track_features[4]])
    
    # use normalized distances to all reference tracks
    ref_df["distances"] = ref_df["song_vec"].apply(lambda x: norm(track_songvec-np.array(x)))
    
    # sort distances from lowest to highest
    ref_df_sorted = ref_df.sort_values(by = "distances", ascending = True)
    
    # return n recommendations
    return ref_df_sorted.iloc[:n_recs]
```

```python
def mood(track_id, df):
    ref_df = df.copy()
    
    # Z-Transform the data
    valence = (ref_df.valence - ref_df.valence.mean()) / ref_df.valence.std()
    energy = (ref_df.energy - ref_df.energy.mean()) / ref_df.energy.std()
    
    moods = pd.concat([valence, energy], axis=1)
    
    # Get Kmeans of the ref_df of songs
    kmeans = KMeans(4, random_state=0)
    kmeans.fit(moods)

    # valence and arousal of given track from spotify api
    track_features = song_features(track_id)
    track_valence, track_energy = [track_features[9], track_features[4]]
    
    # get z score of track
    track_z_val = (track_valence - ref_df.valence.mean()) / ref_df.valence.std()
    track_z_energy = (track_energy - ref_df.energy.mean()) / ref_df.energy.std()
    track = pd.DataFrame([[track_z_val, track_z_energy]], columns=['valence', 'energy'])
    track = moods.append(track)
    
    # predict the cluster mood that the track belongs to
    predicted_cluster = kmeans.fit_predict(track)[-1]

    
    # assign cluster to the mood given by the valence arousal
    if predicted_cluster == 0:
        emotion = 'Relaxed'
    elif predicted_cluster == 1:
        emotion = 'Stressed'
    elif predicted_cluster == 2:
        emotion = 'Sad'
    elif predicted_cluster == 3:
        emotion = 'Happy'
    
    # Return n recommendations
    return emotion
```
